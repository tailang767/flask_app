```python
#単C999円以下の単語.pkl
#単C1000円以上の単語.pkl
#単C2000円以上の単語.pkl
#プロパー5000円以上の単語.pkl
#プロパー10000円以上のタイトル.pkl

import pickle

data = {'zyanru1': ['1', '2', '3', '4', '5'], 'zyanru2': ['kk', 'ss', 'tt']}

with open("単C1000円以上の単語.pkl", "wb") as f:
    pickle.dump(data, f)
```

```python
import pickle

with open("単C1000円以上の単語.pkl", "rb") as f:
    data = pickle.load(f)

print(data)
```
